# **II. Data Cleaning**

## 3.Seperating data



```python
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
data.head(10)
```

## 4.Normalizing values


```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(train_data[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'lat', 'long', 'sqft_living15', 'sqft_lot15']])
```

```python
features_scaled.shape
```

# **IV. Creating and validating a model**


```python
all_features = data[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'lat', 'long', 'sqft_living15', 'sqft_lot15']]
all_features_scaled = scaler.transform(all_features)
```

```python
all_target = data['price']
```

```python
model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
```

```python
model.fit(features_scaled, train_data['price'])
```

```python
all_predictions = model.predict(all_features_scaled)
```

```python
r2 = r2_score(all_target, all_predictions)
```

```python
print(f"Le coefficient of détermination R² is : {r2:.4f}")
```
